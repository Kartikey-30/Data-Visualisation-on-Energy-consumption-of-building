```python
import numpy as np
import pandas as pd
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
whether_data_direc = "K:/new/weather_comparison/new_iiitWdataset.csv"
df_w = pd.read_csv(whether_data_direc)
new_df_w = df_w.drop(['Airport_Humidity','IIITD_Humidity'],axis=1)
new_df_w
new_df_w['timestemp']=pd.to_datetime(new_df_w['timestemp'].astype(str), format='%d-%m-%Y %H:%M')
new_df_w.set_index('timestemp',inplace = True)
plt.style.use('fivethirtyeight')
new_df_w.loc['2018-03-01':'2018-03-15'].plot(figsize=(14,8))
plt.legend(loc ='lower right')
plt.title('temprature comparision plot.')
plt.xlabel('timestamp')
plt.ylabel('temperature')
```
