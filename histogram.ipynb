```python
import numpy as np
import pandas as pd
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
energy_data_direc = "K:/new/energy_dataset/girls_hostel_mains.csv"
df_e = pd.read_csv(energy_data_direc,index_col = 'timestamp')
df_e.index = pd.to_datetime(df_e.index,unit = 's')
df_e.index =  df_e.index + datetime.timedelta(minutes = 60*5 + 30)
new_df_e = df_e.drop(['current','voltage','frequency','power_factor'],axis=1)
def divide(x):
    return x/1000
y = new_df_e['power'].apply(divide)
y
```

```python
# Girls hostel main
plt.style.use('fivethirtyeight')
w = 1
n = math.ceil((y.max() - y.min())/w)
plt.hist(y,bins=n,edgecolor='black')
plt.legend('Power')
plt.title('Power consumtion of girls hostel main.')
plt.xlabel('Power')
plt.ylabel('count')
```

```python
a = new_df_e.loc['2013-08']
a = a['power'].apply(divide)
w = 1
n = math.ceil((a.max() - a.min())/w)
a.hist(bins=n, edgecolor = 'Black')
```
